Made-up Comments from Fake Social Media Page

```python
from splinter import Browser
import os
from bs4 import BeautifulSoup as bs
import requests
import time
```

```python
filepath = os.path.join("CyberPsyche_WebPage", "index.html")
with open(filepath) as file:
    html = file.read()
```

```python
# Create a Beautiful Soup object
soup = bs(html, 'html.parser')
```

```python
# Print formatted version of the soup
print(soup.prettify())
```

```python
# The text of the first paragraph
soup.body.find('p').text
```

```python
soup.body.find_all("div", class_="comment")
```

```python
# Print only the comments
comments = []
for item in soup.body.find_all("div", class_="comment"):
    comments.append(item.p.text)
    
print(comments)
```
